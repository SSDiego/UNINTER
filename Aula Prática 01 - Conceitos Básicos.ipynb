#### Uma lista completa de clientes pode ser definida da seguinte forma
Você deve ter notado que ao definir os vetores de características, não utilizamos o nome do cliente e substituímos os valores de algumas variáveis por valores numéricos. Isso será necessário, pois como conversamos sobre os tipos de atributos e dados, todos valores devem ser convertidos para números.
Os atributos idade, peso e salário, são exemplos de valores contínuos.
Os atributos nacionalidade, sexo e "Gosta de Futebol?", são exemplos de valores categóricos.

O nome do cliente, ainda que seja algo que permita descrevê-lo, não é necessário para os métodos de aprendizagem de máquina. Ainda que seja possível, para abordagens mais avançadas, compararmos os nomes utilizando atributos relacionados a eles.

```python
import pandas as pd

todos_clientes = [[22, 0, 1.70, 75, 0, 1, 2500],
                  [52,1,1.75,80,0,1,2500],
                  [31,1,1.50,65,1,1,4500],
                  [65,1,1.95,86,0,0,6500],
                  [17,0,1.81,95,0,1,1500],
                  [54,0,1.65,80,1,0,3500],
                  [30,0,1.90,105,1,1,1200],
                  [25,1,1.85,65,1,1,2245],
                  [49,0,1.71,75,0,0,25000],
                  [26,1,1.81,80,1,1,8000]]

df_clientes = pd.DataFrame(data=todos_clientes, 
                           columns=["idade", "nacionalidade",
                                    "altura", "peso", "sexo",
                                    "gosta de futebol", "salario"])

df_clientes
```

#### Os vetores de características possibilitam entender melhor os nossos dados

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Aqui criamos uma lista com os valores das nacionalidades
# e associamos cores a elas
nacionalidades = np.unique(["brasileiro", "estrangeiro"])
cores = ['green', 'purple']

# Desenha o gráfico para cada categoria
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')

for i, nac in enumerate(nacionalidades):
    
    idades = [ dados_cliente[0] for dados_cliente in todos_clientes if dados_cliente[1]==i ]
    
    salarios = [dados_cliente[6] for dados_cliente in todos_clientes if dados_cliente[1]==i]
    
    plt.scatter(idades, salarios, s=30, c=cores[i], label=str(nac))
    

# Ajustes no leiaute
plt.gca().set(xlim=(0, 100), ylim=(0, 30000),
              xlabel='Idade', ylabel='Salário')

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Distribuição das Nacionalidades por Idade e Salário", fontsize=18)
plt.legend(fontsize=18)    
plt.show()  
```

#### Aqui encontramos um outlier. Uma instância que se distância de maneira anormal das demais da amostra.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Aqui criamos uma lista com os valores das nacionalidades
# e associamos cores a elas
nacionalidades = np.unique(["brasileiro", "estrangeiro"])
cores = ['green', 'purple']

# Desenha o gráfico para cada categoria
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')

for i, nac in enumerate(nacionalidades):
    idades = [dados_cliente[0] for dados_cliente in todos_clientes if dados_cliente[1]==i]
    salarios = [dados_cliente[6] for dados_cliente in todos_clientes if dados_cliente[1]==i]
    plt.scatter(idades, salarios, s=30, c=cores[i], label=str(nac))
    
# Indica a presença do outlier
plt.annotate('outlier', xy=(49, 25000), xytext=(40, 26000),
             arrowprops=dict(arrowstyle="->",facecolor='black', 
                             connectionstyle="arc3"))    
    
# Ajustes no leiaute
plt.gca().set(xlim=(0, 100), ylim=(0, 30000),
              xlabel='Idade', ylabel='Salário')

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Distribuição das Nacionalidades por Idade e Salário", fontsize=18)
plt.legend(fontsize=18)    
plt.show() 
```

### Exemplo de tratamento não-supervisionado de dados

```python
import pandas as pd

todos_usuarios = [[1,22,0,0,1,4106902,1],
                  [2,30,1,1,1,4105508,0],
                  [3,40,0,1,1,4106902,1],
                  [4,35,0,1,1,4105508,1],
                  [5,26,0,0,1,4106902,0]]

df_usuarios = pd.DataFrame(data=todos_usuarios, 
                           columns=["id", "idade",
                                    "gosta de esportes", "gosta de ler", "gosta de filmes",
                                    "codigo municipio", "sexo"])

df_usuarios
```

### Visualização dos grupos

```python
# Aqui criamos uma lista com os valores das nacionalidades
# e associamos cores a elas
grupos = ["Grupo 01", "Grupo 02"]
cores = ['red', 'green']

# Desenha a distribuição de cada uma das categorias
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')

# Soma os interesses e acrescenta ao final de cada vetor de características
dados_usuarios = []
for i in range(len(todos_usuarios)):
    dados_usuarios.append(todos_usuarios[i]+[todos_usuarios[i][2]+todos_usuarios[i][3]+todos_usuarios[i][4]])  
                                        
idades_g1 = [dados_usuario[1] for dados_usuario in dados_usuarios if dados_usuario[7] < 2]    
interesses_g1 = [dados_usuario[7] for dados_usuario in dados_usuarios if dados_usuario[7] < 2] 

idades_g2 = [dados_usuario[1] for dados_usuario in dados_usuarios if dados_usuario[7] >= 2]    
interesses_g2 = [dados_usuario[7] for dados_usuario in dados_usuarios if dados_usuario[7] >= 2]    

plt.scatter(idades_g1, interesses_g1, s=50, c=cores[0], label=str(grupos[0]))
plt.scatter(idades_g2, interesses_g2, s=50, c=cores[1], label=str(grupos[1]))

for i,dados_usuario in enumerate(dados_usuarios):
    x = dados_usuario[1]
    y = dados_usuario[7]
    label = "id {0}".format(dados_usuario[0])
    plt.annotate(label, 
                 (x,y), 
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center') 
    
# Legendas
plt.gca().set(xlim=(0, 80), ylim=(0, 5))

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Agrupamentos de Usuários", fontsize=18)
plt.legend(fontsize=18)    
plt.show() 
```
