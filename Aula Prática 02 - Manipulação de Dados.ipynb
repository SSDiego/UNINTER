```python
# Convertendo o dataframe pandas para xml
def to_xml(df, filename=None, mode='w'):
    def row_to_xml(row):
        xml = ['<cliente>']
        for i, col_name in enumerate(row.index):
            xml.append('  <atributo name="{0}">{1}</atributo>'.format(col_name, row.iloc[i]))
        xml.append('</cliente>')
        return '\n'.join(xml)
    res = '\n'.join(df.apply(row_to_xml, axis=1))

    if filename is None:
        return res
    with open(filename, mode) as f:
        f.write(res)

pd.DataFrame.to_xml = to_xml

df_2clientes = pd.DataFrame([cliente_01,cliente_02])
str_xml = df_2clientes.to_xml()
print(str(str_xml))
```
